```python
def smart_truncate(content, length=200, suffix='...'):
    if len(content) <= length:
        return content
    else:
        return ' '.join(content[:length+1].split(' ')[0:-1]) + suffix

appendix = model.topic_names[model.topic_names['topic'] != -1].copy()
representative_companies = []
for i in range(197):
    representative_company = model.topic_model.get_representative_docs(i)[0]
    representative_companies.append(representative_company)
appendix['representative_company'] = representative_companies
appendix['representative_company'] = appendix['representative_company'].apply(smart_truncate)
appendix.to_csv('appendix.csv', index=False, encoding='utf-8')
```
